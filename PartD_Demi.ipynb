```python
import pandas as pd

import nltk
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords


nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer


```

```python
train_set = pd.read_csv("climate_train.csv")
test_set = pd.read_csv("climate_test.csv")
```

```python
tokenizer = RegexpTokenizer(r'\w+')
lemmatizer = WordNetLemmatizer()
stopwords = set(stopwords.words('english'))

def preprocessing(text):
    tokens = tokenizer.tokenize(text)
    after_stopwords = [word.lower() for word in tokens if word.lower() not in stopwords] 
    lemmas = [lemmatizer.lemmatize(word) for word in after_stopwords]

    return lemmas
```

```python
test_set
```

```python
preprocessed_train_set = pd.DataFrame(train_set["text"].apply(preprocessing))
preprocessed_test_set = pd.DataFrame(test_set["text"].apply(preprocessing))

preprocessed_train_set['label'] = train_set['label']
preprocessed_test_set['label'] = test_set['label']
```
